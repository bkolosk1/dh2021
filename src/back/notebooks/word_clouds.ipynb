```python
import pickle
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
with open('text_data.pkl','rb') as f:
    data = pickle.load(f)
```

```python
for x in data:
    concated = " ".join(data[x])
    wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(concated)
    plt.figure()
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.show()
    plt.savefig('../../data/imgs/'+x+".png")
```
